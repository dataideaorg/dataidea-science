**Week 3: Linear Regression in PyTorch**

1. **Stochastic Gradient Descent:**
   - **Exercise:** Implement a linear regression model in PyTorch and train it using Stochastic Gradient Descent (SGD). Compare the performance with a model trained using regular gradient descent.

2. **Mini-Batch Gradient Descent:**
   - **Exercise:** Modify your PyTorch implementation to use Mini-Batch Gradient Descent. Train the model and evaluate its performance using different batch sizes.

3. **PyTorch Built-in Functions:**
   - **Exercise:** Utilize PyTorch’s built-in functions (`torch.optim` and `torch.nn`) to implement and train a linear regression model. Compare the results with your manual implementation.

4. **Training and Validation Sets:**
   - **Exercise:** Split a dataset into training and validation sets. Train a linear regression model on the training set and evaluate its performance on the validation set. Report the loss and any metrics of interest.
