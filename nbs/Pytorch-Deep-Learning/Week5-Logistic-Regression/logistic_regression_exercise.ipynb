**Week 5: Logistic Regression**

1. **Making Predictions in Logistic Regression:**
   - **Exercise:** Implement logistic regression using PyTorch. Train the model on a binary classification dataset and make predictions on new data. Evaluate the model using accuracy.

2. **Logistic Regression and Bad Initialization Values:**
   - **Exercise:** Train a logistic regression model with various initialization strategies. Compare the impact of different initial values on the model’s convergence and performance.

3. **Cross Entropy Loss Function:**
   - **Exercise:** Implement a logistic regression model using PyTorch with Cross Entropy Loss. Compare the performance with a model using Mean Squared Error (MSE) Loss for the same classification problem.

4. **Softmax Activation in 1 Dimension:**
   - **Exercise:** Implement the softmax function from scratch and apply it to a vector of raw scores. Verify that the output probabilities sum to 1 and plot the softmax activation for different input values.
