**Week 4: Multiple Input Linear Regression**

1. **Making Predictions in Multiple Linear Regression:**
   - **Exercise:** Implement a multiple linear regression model using PyTorch with at least 3 features. Train the model on a synthetic dataset and make predictions.

2. **Training a Multiple Linear Regression Model:**
   - **Exercise:** Create and train a multiple linear regression model using real-world data (e.g., housing prices). Evaluate the model's performance using metrics such as Mean Absolute Error (MAE).

3. **Multi-Target Linear Regression:**
   - **Exercise:** Extend your multiple linear regression model to handle multiple target variables. Train the model on a synthetic dataset with two target variables and evaluate its performance.

4. **Training Multiple Output Linear Regression Models:**
   - **Exercise:** Implement and train a model that can predict multiple outputs simultaneously. Use a dataset with multiple features and multiple target variables.
