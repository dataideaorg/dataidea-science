**Week 2: Linear Regression**

1. **Linear Regression 1 Dimension:**
   - **Exercise:** Implement a linear regression model for a 1D dataset using only numpy. Create synthetic data with a linear relationship and fit a model to this data. Plot the data and the fitted line.

2. **Linear Regression with 1 Parameter:**
   - **Exercise:** Using scikit-learn, perform linear regression on a dataset with a single feature and one parameter. Compare the results with the numpy implementation from the previous exercise. 

3. **Training Slope and Bias:**
   - **Exercise:** Create a function to compute the slope and bias of a linear regression model manually using gradient descent. Use synthetic data to test your implementation.
