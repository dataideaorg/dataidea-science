## **Question 1:** Which type of chart is used to show the relationship between two continuous variables?

a) Bar chart  
b) Pie chart  
c) Scatter plot  
d) Area chart  

## **Question 2:** What is the primary purpose of a radar chart?

a) Comparing multiple categories across a few variables  
b) Showing hierarchical data  
c) Displaying trends over time  
d) Visualizing correlations between two variables  

## **Question 3:** In a stacked bar chart, each bar is divided into segments to represent:

a) Different categories of a single variable  
b) Time periods  
c) Trends over time  
d) Correlation between variables  

## **Question 4:** When should a heat map be used?

a) Showing proportions within a population  
b) Comparing individual data points  
c) Visualizing spatial relationships and correlations  
d) Displaying data distribution  

## **Question 5:** Which type of chart is best suited for comparing parts of a whole?

a) Scatter plot  
b) Histogram  
c) Pie chart  
d) Line chart  

## **Question 6:** Which chart type is used to detect outliers and visualize the distribution of a dataset?

a) Box plot  
b) Area chart  
c) Bubble chart  
d) Radar chart  

## **Question 7:** What does a bubble chart represent?

a) Relationship between two continuous variables  
b) Relationship between two categorical variables  
c) Relationship between three variables using two continuous variables and one categorical variable  
d) Relationship between three continuous variables  

## **Question 8:** Which chart type is suitable for showing hierarchical data with nested categories?

a) Tree map  
b) Scatter plot  
c) Line chart  
d) Histogram  

## **Question 9:** When is a bar chart more appropriate than a pie chart?

a) Comparing parts of a whole  
b) Showing trends over time  
c) Displaying a single continuous variable  
d) Visualizing correlations  

## **Question 10:** What is the main purpose of a histogram?

a) Showing hierarchical data  
b) Visualizing correlations  
c) Displaying the distribution of a single continuous variable  
d) Comparing multiple categories across variables  

---

## **Answers:**

1. c) Scatter plot
2. a) Comparing multiple categories across a few variables
3. a) Different categories of a single variable
4. c) Visualizing spatial relationships and correlations
5. c) Pie chart
6. a) Box plot
7. c) Relationship between three variables using two continuous variables and one categorical variable
8. a) Tree map
9. a) Comparing parts of a whole
10. c) Displaying the distribution of a single continuous variable
