7. A car increases it velocity from u ms-1 to v ms-1 within t seconds. Write a program to
calculate the acceleration.

````python
# your solution

u = float(input("Enter the initial velocity (m/s): "))
v = float(input("Enter the final velocity (m/s): "))
t = float(input("Enter the time (s): "))

getAcceleration = lambda u, v, t: (v - u) / t

print("The acceleration is:", getAcceleration(u, v, t), "m/s^2")
````

8. Write a program to input a temperature reading in either Celsius(c) or Fahrenheit(f)
scale and convert it to the other scale. The temperature reading consists of a decimal number
followed by letter ”F” or ”f” if the temperature is in Fahrenheit scale or letter ”C” or ”c” if it is in
Celsius scale. You may use a similar format for the output of the program.
Note: `c = 5( f − 32) / 9`

````python
# your solution

temperature_reading = input("Enter the temperature: ")

unit = str(temperature_reading[-1])
temperature_value = float(temperature_reading[:-1])

if unit.lower() == "c":
    fahrenheit = (temperature_value * 9/5) + 32
    print(f"{temperature_value}°C is equivalent to {round(fahrenheit, 3)}°F")
else:
    celsius = (temperature_value - 32) * 5/9
    print(f"{temperature_value}°F is equivalent to {round(celsius, 3)}°C")
````

13. Write a program to calculate the factorial of any given positive integer.

````python
# your solution
# factorial
def getFactorial(n):
    if n == 0:
        return 1
    else:
        return getFactorial(n - 1) * n
    
print(getFactorial(5))
````

14. Write a program to compute the sum of all integers between any given two numbers. In this program both inputs should be given from the keyboard.

````python
# your solution

def getSumBetween(start, end):
    sum = 0
    for counter in range(start, end+1):
        sum += counter
    return sum

print(getSumBetween(1, 10)) # 55
````

19. Write a program to store marks of 5 students for 5 subjects given through the keyboard.
Calculate the average of each students marks and the average of marks taken by all the students

````python

# Write a program to store marks of 5 students for 5 subjects given through 
# the keyboard. Calculate the average of each students marks and the average 
# of marks taken by all the students

# your solution
student_marks = []
for i in range(5):
    student_marks.append([])
    for j in range(5):
        mark = int(input(f"Enter marks for student {i+1}, subject {j+1}: "))
        student_marks[i].append(mark)

for i in range(5):
    student_average = sum(student_marks[i]) / len(student_marks[i])
    print(f"Average marks for student {i+1}: {student_average:.2f}")

all_marks = []
for student in student_marks:
    all_marks.extend(student)

overall_average = sum(all_marks) / len(all_marks)
print(f"Overall average of all students: {overall_average:.2f}")
````
